```python
features= ["overall", "potential", "wage_eur", "value_eur", "age"] # clustering on basis of only these 5 features
players= players.dropna(subset=features) # dropping all rows having zero values in any of above mentioned 5 columns
data= players[features].copy() #  copying only 5 columns of player data into another dataframe
data
```

```python
# normalizing values in all 5 columns for avoiding issues in scaling
data = ((data- data.min()) /(data.max()- data.min()))* 10 + 1 # scales value on a scale of 10 
data.describe()
```

```python
def random_centroids(data, k):
    centroids = []  # creating empty list of centroids to be stored
    for i in range(k): 
        centroid = data.apply(lambda x: float(x.sample()))  # in each iteration, centroid will take value at random from data via lambda function and convert to float 
        centroids.append(centroid)  # this value will then be appended to the centroids list
    return pd.concat(centroids, axis=1)  # returning a dataframe containing all centroids in columns (axis 1)
```

```python
centroids= random_centroids(data, 5)
```

```python
centroids
```

```python
def get_labels(data, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
    return distances.idxmin(axis=1)
```

```python
labels = get_labels(data, centroids)
```

```python
labels.value_counts()
```

```python
def new_centroids(data, labels, k):
    centroids = data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
    return centroids
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels)
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1])
    plt.show()
```

```python
max_iterations = 100
centroid_count = 3

centroids = random_centroids(data, centroid_count)
old_centroids = pd.DataFrame()
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids
    
    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, centroid_count)
    plot_clusters(data, labels, centroids, iteration)
    iteration += 1
```
